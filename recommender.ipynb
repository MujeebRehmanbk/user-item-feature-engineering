```python
import pandas as pd
import numpy as np
import random
```

```python

def data_preparation(all_data):
    
    global m
    global mv
    global tv
    global mov
    global live
    
    m   =  list(all_data[all_data["content_map"] == "music"]["title"].values)
    mv  = list(all_data[all_data["content_map"] == "music_video"]["title"].values)
    tv  = list(all_data[all_data["content_map"] == "TV"]["title"].values)
    mov = list(all_data[all_data["content_map"] == "movie"]["title"].values)
    live= list(all_data[all_data["content_map"] == "live"]["title"].values)


def user_profiling(dataframe,n):
    
    user_movies = []
    user_tv = []
    user_live = []
    user_music = []
    user_music_videos = []

    for i in range(n):
        u_movies = random.choice(mov).strip()
        u_tv = random.choice(tv).strip()
        u_music = random.choice(m).strip()
        u_music_videos = random.choice(mv).strip()
        u_live = random.choice(live).strip()

        user_movies.append(u_movies)
        user_tv.append(u_tv)
        user_live.append(u_live)
        user_music.append(u_music)
        user_music_videos.append(u_music_videos)
        
    user_profiles = list(set(user_movies + user_tv + user_live + user_music + user_music_videos))
    
    user_genres  = list(all_data[all_data["title"].isin(user_profiles)]['genre_name'].values)
    user_genres_count = []
    for i in set(user_genres):
        count = user_genres.count(i)
        user_genres_count.append((i,count))
    s =[]
    for i in set(user_genres):
        bolean_map = ~all_data["title"].isin(user_profiles)
        a = all_data[bolean_map]
        a = all_data[all_data["genre_name"] == i]\
        .sort_values(by = ["times_played"],ascending = False)[["id","title","genre_name","content_map"]]\
        .values.tolist()[:5]
        s = s+a
    return s

def recommender(profile,n):
    recomendations = []
    i = len(recomendations)
    while i < n:
        recomend = random.choice(list(profile))
        if recomend not in recomendations:
            recomendations.append(recomend)
        i = len(recomendations)
    df = pd.DataFrame(recomendations,columns= ["id","title","genre","content_type"])
    return df

```
