```python
### video_genres -> movies, TV
video_content = {1:"movie",
                2:"TV",
                3:"music_video",
                4:"live"}

video["content_map"] = video.replace({"content_type":video_content})["content_type"]

video_cols = ["id","title","times_played","content_map","content_type","genre_id"]

video =  video[video_cols].sort_values(by = ["content_map","times_played"],ascending= False)
video.head()
```

```python
### video_genres -> movies, TV
tv_movies = video[(video["content_map"] == "movie") | (video["content_map"] == "TV")]
tv_movies.head()
```

```python
tv_movies_map = tv_movies.merge(video_genres, on = "genre_id")[video_cols +["genre_name"]]
tv_movies_map.head()
```

```python
## live, music_video -> music genre
## podcasts -> ask gokhan

live_music = video[(video["content_map"] == "live") | (video["content_map"] == "music_video")]
live_music.head()
```

```python
live_music_map = live_music.merge(music_genres, on = "genre_id").sort_values(by = ["content_map","times_played"],
                                                                            ascending = False)[video_cols+["genre_name"]]
live_music_map.head()
```

```python
music_col = ["id","title","genre_id","times_played","content_map","content_type"]
music['content_map'] = ["music"]*music.shape[0]
music['content_type'] = [0]*music.shape[0]

music_map = music[music_col].merge(music_genres, on = "genre_id").sort_values(by = ["content_map","times_played"],
                                                                    ascending = False)[video_cols+["genre_name"]]
music_map.head()
```

```python
all_data = pd.concat([live_music_map, tv_movies_map, music_map])
print(all_data.shape)
all_data.head()

```

```python
m = list(all_data[all_data["content_map"] == "music"]["title"].values)
mv = list(all_data[all_data["content_map"] == "music_video"]["title"].values)
tv = list(all_data[all_data["content_map"] == "TV"]["title"].values)
mov = list(all_data[all_data["content_map"] == "movie"]["title"].values)
live = list(all_data[all_data["content_map"] == "live"]["title"].values)
```
